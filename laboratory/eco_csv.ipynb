```python
import pandas as pd
```

```python
excel_file = pd.read_excel('data/economic_data.xlsx')
excel_file
```

```python
excel_file.country.value_counts().head()
```

```python
excel_file_filter = excel_file[excel_file['country'].isin(['US', 'JP', 'GB', 'DE', 'CA', 'EU', 'AU', 'CN', 'CH', 'NZ'])]
excel_file_filter
```

```python
excel_file_cp = excel_file.copy()
excel_file_cp['comment'] = excel_file_cp['comment'].fillna('')
```

```python
switzerland_rows = excel_file_cp[excel_file_cp['comment'].str.contains('Switzerland', case=False)]
switzerland_rows
```

```python
newzealand_rows = excel_file_cp[excel_file_cp['comment'].str.contains('New Zealand', case=False)]
newzealand_rows
```

```python
csv_file = excel_file_filter.to_csv('data/economic_data.csv', index=False)
```
