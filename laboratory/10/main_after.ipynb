```python
sr = filter_quote.loc[:, ('Short_Distance', 'Long_Distance', 'Relative_SL', 'RelativePrice_SL')].mean()
sr['Combine_Rel'] = sr['Short_Distance'] / sr['Long_Distance']
sr
```

```python
# sr['Combine_Rel']
```

```python
filter_quote['Bed_av'] = filter_quote.Short_Distance.div(sr['Short_Distance'])
filter_quote['Bud_av'] = filter_quote.Long_Distance.div(sr['Long_Distance'])
filter_quote['Rel_avg'] = filter_quote.Relative_SL.div(sr['Relative_SL'])

filter_quote['Rel_com_avg'] = filter_quote.Relative_SL.div(sr['Combine_Rel'])

filter_quote['Rel_p_avg'] = filter_quote.RelativePrice_SL.div(sr['RelativePrice_SL'])
```

```python
filter_quote = filter_quote.head(10)
```
