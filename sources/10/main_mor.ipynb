```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20,10)
pd.set_option('max_columns', 15, 'max_rows', 30, 'display.precision', 4)
```

```python
result = pd.read_csv('../cookbook_data/owndata/prices_cleandf2202.csv', index_col=0)
result_cmp = result.copy()
```

##### Notice: GBPJPY, GBPCAD, EURJPY, USDJPY, AUDJPY, NZDJPY

```python
quote = r'GBP'
filter_quote = result_cmp[result_cmp.index.str.contains(quote + r'(?!$)')]
filter_quote = filter_quote.apply(pd.to_numeric)
```

```python
sr = filter_quote.loc[:, ('Short_Distance', 'Long_Distance', 'Relative_SL', 'RelativePrice_SL')].mean()
sr['Combine_Rel'] = sr['Short_Distance'] / sr['Long_Distance']
```

```python
sr['Combine_Rel']
```

```python
filter_quote['Bed_av'] = filter_quote.Short_Distance.div(sr['Short_Distance'])
filter_quote['Bud_av'] = filter_quote.Long_Distance.div(sr['Long_Distance'])
filter_quote['Rel_avg'] = filter_quote.Relative_SL.div(sr['Relative_SL'])

filter_quote['Rel_com_avg'] = filter_quote.Relative_SL.div(sr['Combine_Rel'])

filter_quote['Rel_p_avg'] = filter_quote.RelativePrice_SL.div(sr['RelativePrice_SL'])
```

```python
filter_quote.sort_values(by='Bed_av', ascending=False, key=abs)
```
